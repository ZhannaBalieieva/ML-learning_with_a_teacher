```python
df_1 = pd.read_csv('result.csv')
df_1
print(df_1.describe())
list(df_1.columns)
pd.isnull(df_1)
a = df_1.fillna(0)
a.isnull().sum()
```

```python
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score

'''Dividing the dataset into training and control samples'''

a_shuffled = shuffle(a, random_state=42)
X = a_shuffled[['accelerometer_X', 'accelerometer_Z']]
y = a_shuffled["accelerometer_Y"]
y=y.astype('int')
print(X, y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=85)
ss = StandardScaler()
y=y.astype('int')
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
y=y.astype('int')
y_train = np.array(y_train)
X_train['time_mean'] = X_train_scaled.mean(axis=1)
X_test['time_mean'] = X_test.mean(axis=1)

```

```python
from sklearn.model_selection import cross_val_score

'''Support Vector Classifier'''
SVC_model = svm.SVC()
SVC_model.fit(X_train, y_train)
scores = cross_val_score(SVC_model,X_train, y_train)
scores.mean()
```

```python
rfc_model = RandomForestClassifier()
rfc_model.fit(X_train, y_train)
```

```python
SVC_prediction = SVC_model.predict(X_test)
rfc_prediction = rfc_model.predict(X_test)

```

```python
print(accuracy_score(SVC_prediction, y_test))
print(accuracy_score(rfc_prediction, y_test))
```

```python
from sklearn.metrics import classification_report

svm_report = classification_report(y_test, SVC_prediction)
print("SVM Results:")
print(svm_report)

rf_report = classification_report(y_test, rfc_prediction)
print("Random Forest Results:")
print(rf_report)
```
